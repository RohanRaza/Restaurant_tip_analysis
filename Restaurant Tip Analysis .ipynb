```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("tips.csv")
data.head()
```

```python
data.shape
```

## Checking relation between Toatal bill to tip

```python
sns.scatterplot(x = "total_bill", y = "tip", data = data)
sns.despine()

plt.show()
```

### Corelation

```python
data.corr()["tip"]['total_bill']
```

```python
sns.heatmap(data.corr(), annot = True)

plt.show()
```

## Result: Tip is directly proportional to Total Bill

## Checking Relation between Tip to Family Size

```python
sns.scatterplot(x = 'size', y = 'tip', data = data)


plt.show()
```

```python
data['size'].value_counts()
```

```python
data['tip'].describe()
```

```python
sns.set_palette('bright')

sns.regplot(x = 'size', y = 'tip',data = data)

sns.despine()

plt.show()
```

```python
data.corr()['tip']['size']
```

```python
sns.heatmap(data.corr(),annot = True, cmap = 'crest')

plt.show()
```

## Result: Size has moderate relation with Tip

## Reatioin between tip and day

```python
sns.set_style('dark')

sns.scatterplot(x = 'day', y = 'tip', data = data)
sns.despine(offset = 10, trim = True)

plt.show()
```

```python
data['day'].value_counts()
```

### Result: On sat tip ratio is high

#### Since tip is quantitative and day is qualitative so we can't find co-relation

```python
sns.set_palette('bright')
sns.set_context('poster')
sns.scatterplot(x = 'time', y = 'tip', data = data)
sns.despine()
sns.despine(offset = 10, trim = True )

plt.show()
```

## Result: On dinner time tip ratio is high

## Now we see all atributes in sub plots by using Relplot

```python
g = sns.relplot(x = 'total_bill', y = 'tip', data = data, hue = 'sex', size = 'day', style = 'smoker', col = 'time')
g.fig.set_figheight(8)
g.fig.set_figwidth(12)

plt.show()
```

## Final Result
### Tip has directly relation with 'total bill', Tip is more on 'Sat Night', max tip is given by 'Male' who was as 'smoker', Tip has moderate relation with family size.
